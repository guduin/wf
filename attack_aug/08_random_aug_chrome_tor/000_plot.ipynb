```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
res_tcp_ramdom = np.load('data/res_tcp_ramdom.npy')
res_tcp_ramdom_aug = np.load('data/res_tcp_ramdom_aug.npy')
res_tcp_mtu = np.load('data/res_tcp_mtu.npy')
res_tcp_mtu_aug = np.load('data/res_tcp_mtu_aug.npy')
res_tcp_split = np.load('data/res_tcp_split.npy')
res_tcp_split_aug = np.load('data/res_tcp_split_aug.npy')
```

```python
plt.figure(figsize=(6,5))

x = np.linspace(0, 1, num=21, endpoint=True)

plt.plot(x, res_tcp_ramdom, 'ro-')
plt.plot(x, res_tcp_ramdom_aug, 'r^--')
plt.plot(x, res_tcp_mtu, 'go-')
plt.plot(x, res_tcp_mtu_aug, 'g^--')
plt.plot(x, res_tcp_split, 'bo-')
plt.plot(x, res_tcp_split_aug, 'b^--')

plt.xlabel('Random Packets Probability', fontsize=18)
plt.ylabel('WF Attack Accuracy', fontsize=18)
plt.xlim((0, 1))
plt.ylim((0, 1))
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(['RPD Insert', '3A Insert', 'RPD MTU', '3A MTU', 'RPD Split', '3A Split'], loc=4)
plt.savefig('C:/Users/Admin/Desktop/random_aug_chrome.pdf', bbox_inches = 'tight')
```
