```python
train_data_edge, train_index_edge, test_data_edge, test_index_edge = module.load_data('F:/data_edge/3_data/data.csv')
train_data_chrome, train_index_chrome, test_data_chrome, test_index_chrome = module.load_data('F:/data_chrome/3_data/data.csv')
train_data_firefox, train_index_firefox, test_data_firefox, test_index_firefox = module.load_data('F:/data_firefox/3_data/data.csv')
```

```python
stat = clt.Counter(train_data_edge.flatten())
del(stat[0])
count_pkts = sum(stat.values())
stat_sort = np.array(sorted(stat.items(), key=lambda x:x[0]))

size = stat_sort[:, 0]
prob = stat_sort[:, 1] / count_pkts

x = np.linspace(-1500, 1500, num=61)
y = np.zeros(61)

for i in range(len(size)):
    for j in range(len(x)):
        if size[i] > x[j]-25 and size[i] <= x[j]+25:
            y[j] += prob[i]
```

```python
plt.figure(figsize=(6,5))
plt.bar(x, y, width=40, zorder=100, color='b')
plt.ylim([0, 0.1])
plt.xlabel('Packet Size (bin size = 50)', fontsize=18)
plt.ylabel('Packet Probability', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid()
plt.text(-1450, 0.095, str(round(y[0], 2)), fontsize=14)
plt.text(100, 0.095, str(round(y[31], 2)), fontsize=14)
plt.savefig('C:/Users/Admin/Desktop/size_edge.pdf', bbox_inches = 'tight')
```
